```python
import matplotlib.pyplot as plt 

x=[1,2,3,4,5,]
y=[10,15,20,35,50]

plt.plot(x,y)
plt.title("Attendees")
plt.xlabel("X-axis")
```

```python
names = ['James', 'Cane', 'Evans', 'Matthew']
marks = [56, 78, 67, 98]

plt.bar(names,marks, color='blue')
plt.xlabel('Names')
plt.ylabel('Marks')
```

```python
activities = ['Coding', 'Gaming', 'Studying', 'Friends']
hours = [4, 2, 5, 3]

plt.pie(hours, labels=activities)
```
